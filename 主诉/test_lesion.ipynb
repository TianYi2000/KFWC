```python
def select_side (alls, parts, pos):
    alls_side = [all for all in alls if all['pos'] == pos ]
    parts_side = [part for part in parts if pos in part['text'] or '双眼' in part['text']]
    alls_side = sorted(alls_side, key=operator.itemgetter('time'))
    parts_side = sorted(parts_side, key=operator.itemgetter('time'))
    return alls_side, parts_side

def distri (alls, parts, name):
    if len(alls) == 0 or len(parts) == 0:
        return None
    split_len = int(len(alls) // len(parts))
    split_nums = [split_len] * len(parts)
    rest = len(alls) - split_len * len(parts)
    for i in range(rest):
        split_nums[i] += 1
    now = 0
    NEW_TEXT = []
    for index, part in enumerate(parts):
        start = now
        end = now + split_nums[index]
        for now in range(start, end):
            all = alls[now]
            newline = [all['url'], all['disease']] + all['lesion'] + [name, all['pos'], part['text'], part['time']]
            NEW_TEXT.append(newline)
    return NEW_TEXT
```

```python
NEW_TEXT = []
import operator
for name in NAMES:
    if name in ALL:
        parts = PART[name]
        alls = ALL[name]
        for pos in ['左眼', '右眼', '无法区分左右眼']:
            alls_side, parts_side = select_side(alls, parts, pos)
            new_lines = distri(alls_side, parts_side, name)
            if new_lines != None:
                NEW_TEXT += new_lines
print(NEW_TEXT)
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('./OCT图像-疾病-体征-主诉.csv',encoding='utf-8')
all = np.array(data)
print(all[0])
```

```python
ALL = {
    'CNV' : {'total' : 0},
    'PCV' : {'total' : 0},
    'Non-wet-AMD' : {'total' : 0},
}
for line in all:
    line = line.tolist()
    url = line[0]
    disease = line[1]
    lesion = line[2:12]
    name = line[12]
    pos = line[13]
    text = line[14]
    time = line[15]
    if name not in ALL[disease]:
        ALL[disease][name] = []
    ALL[disease][name].append({
        'url' : url,
        'disease' : disease,
        'lesion' : lesion,
        'name' : name,
        'pos' : pos,
        'text' : text,
        'time' : time,
    })
    ALL[disease]['total'] += 1
# print(ALL)

import random

NEW_ALL = {
    'train' : [],
    'val' : [],
    'test' : [],
}

for disease in ['CNV', 'PCV', 'Non-wet-AMD']:
    TEMP = {
        'train' : [],
        'val' : [],
        'test' : [],
    }

    alls = ALL[disease]
    total = alls['total']
    del alls['total']
    train_nums = int(0.78 * total)
    val_nums = int(0.1 * total)
    test_nums = int(0.1 * total)

    NUMS = {
        'train' : train_nums,
        'val' : val_nums,
        'test' : test_nums,
        'val-test' : 9999999,
    }
    NAMES = list(alls.keys())
    random.shuffle(NAMES)
    val_test = 0
    select = []
    for name in NAMES:
        for phase in [ 'train', 'val-test']:
            if phase != 'train' or len(TEMP[phase]) <= NUMS[phase]:
                for line in alls[name]:
                    newline = [line['url'], line['disease']] + line['lesion'] + [name, line['pos'], line['text'], line['time']]
                    if select != []:
                        phase = select[val_test]
                        val_test = 1 - val_test
                    elif '-' in phase:
                        select = phase.split('-')
                        phase = select[val_test]
                        val_test = 1 - val_test
                    TEMP[phase] .append(newline)
                break
    for phase in ['train', 'val',  'test']:
        NEW_ALL[phase] += TEMP[phase]
# print(NEW_ALL)

DISEASE_TOTAL = {
    'CNV' : 0,
    'PCV': 0,
    'Non-wet-AMD' : 0,
}
print('CNV',
        'PCV',
        'Non-wet-AMD',)
for phase in [ 'train', 'val', 'test']:
    DISEASE_TEMP = {
        'CNV' : 0,
        'PCV': 0,
        'Non-wet-AMD' : 0,
    }
    print(phase, end='')
    for line in NEW_ALL[phase]:
        DISEASE_TEMP[line[1]] += 1
    print(' : ', DISEASE_TEMP['CNV'], DISEASE_TEMP['PCV'], DISEASE_TEMP['Non-wet-AMD'])
    DISEASE_TOTAL['CNV']+= DISEASE_TEMP['CNV']
    DISEASE_TOTAL['PCV']+= DISEASE_TEMP['PCV']
    DISEASE_TOTAL['Non-wet-AMD']+= DISEASE_TEMP['Non-wet-AMD']
print('total : ', DISEASE_TOTAL['CNV'], DISEASE_TOTAL['PCV'], DISEASE_TOTAL['Non-wet-AMD'])
```

```python
for phase in [ 'train', 'val','test']:
    save = pd.DataFrame(NEW_ALL[phase], columns = ['url', 'disease', '视网膜内液性暗腔','视网膜下积液','RPE脱离','RPE下高反射病灶','视网膜内或视网膜下高反射病灶','尖锐的RPED峰','双层征','多发性RPED','RPED切迹','视网膜内高反射硬性渗出','patient_id', 'eye_pos', 'complaints', 'prescribe_time'])
    save.to_csv(f'saved-OCT图像-疾病-体征-主诉/{phase}_label.csv',index=False, encoding='utf-8')  #index=False,
```
