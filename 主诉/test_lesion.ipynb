```python
filename = '所有主诉_新.xlsx'
wb = load_workbook(filename)
ws = wb.get_sheet_by_name('Sheet1')
rows = ws.rows
columns = ws.columns
rows = [row for row in rows]
print(len(rows))
TEXTS = []
for row in rows:
    line = [col.value for col in row]
    TEXTS.append(line)
    # print(line)
print(TEXTS)
```

```python

PART = {}
ID2NAME = {}
for line in TEXTS[1:]:
    name = line[0]
    text = line[2]
    time = line[3]
    id = str(line[4])
    if name not in PART:
        PART[name] = []
    if id != None:
        ID2NAME[id] = name
    PART[name].append({
        'text' : text,
        'time' : time.strftime('%Y-%m-%d')
    })
print(PART)
print(ID2NAME)
```

```python
import random
def distri (alls, parts, name):
    if len(alls) == 0 or len(parts) == 0:
        return None
    NEW_TEXT = []
    for all in alls:
        choose = random.choice(parts)
        if all['disease'] == 'Non-wet-AMD':
            newline = [all['url'], all['disease']] + all['lesion'] + [name, all['pos'], '不能判断为湿性AMD', '00-00-00']
        else:
            newline = [all['url'], all['disease']] + all['lesion'] + [name, all['pos'], choose['text'], choose['time']]
        NEW_TEXT.append(newline)
    return NEW_TEXT
```

```python
NEW_TEXT = []
import operator
for name in NAMES:
    if name in ALL:
        parts = PART[name]
        alls = ALL[name]
        for pos in ['左眼', '右眼', '无法区分左右眼']:
            alls_side, parts_side = select_side(alls, parts, pos)
            new_lines = distri(alls_side, parts_side, name)
            if new_lines != None:
                NEW_TEXT += new_lines
print(len(NEW_TEXT))
print(NEW_TEXT[:5])
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('./OCT图像-疾病-体征-主诉.csv',encoding='utf-8')
all = np.array(data)
print(all[0])
```

```python
ALL = {
    'CNV' : {'total' : 0},
    'PCV' : {'total' : 0},
    'Non-wet-AMD' : {'total' : 0},
}
for line in all:
    line = line.tolist()
    url = line[0]
    disease = line[1]
    lesion = line[2:12]
    name = line[12]
    pos = line[13]
    text = line[14]
    time = line[15]
    if name not in ALL[disease]:
        ALL[disease][name] = []
    ALL[disease][name].append({
        'url' : url,
        'disease' : disease,
        'lesion' : lesion,
        'name' : name,
        'pos' : pos,
        'text' : text,
        'time' : time,
    })
    ALL[disease]['total'] += 1
# print(ALL)

import random

NEW_ALL = {
    'train' : [],
    'val' : [],
    'test' : [],
}

for disease in ['CNV', 'PCV', 'Non-wet-AMD']:
    TEMP = {
        'train' : [],
        'val' : [],
        'test' : [],
    }

    alls = ALL[disease]
    total = alls['total']
    del alls['total']
    train_nums = int(0.78 * total)
    val_nums = int(0.1 * total)
    test_nums = int(0.1 * total)

    NUMS = {
        'train' : train_nums,
        'val' : val_nums,
        'test' : test_nums,
        'val-test' : 9999999,
    }
    NAMES = list(alls.keys())
    random.shuffle(NAMES)
    val_test = 0
    select = []
    for name in NAMES:
        for phase in [ 'train', 'val-test']:
            if phase != 'train' or len(TEMP[phase]) <= NUMS[phase]:
                for line in alls[name]:
                    newline = [line['url'], line['disease']] + line['lesion'] + [name, line['pos'], line['text'], line['time']]
                    if select != []:
                        phase = select[val_test]
                        val_test = 1 - val_test
                    elif '-' in phase:
                        select = phase.split('-')
                        phase = select[val_test]
                        val_test = 1 - val_test
                    TEMP[phase] .append(newline)
                break
    for phase in ['train', 'val',  'test']:
        NEW_ALL[phase] += TEMP[phase]
# print(NEW_ALL)

DISEASE_TOTAL = {
    'CNV' : 0,
    'PCV': 0,
    'Non-wet-AMD' : 0,
}
print('CNV',
        'PCV',
        'Non-wet-AMD',)
for phase in [ 'train', 'val', 'test']:
    DISEASE_TEMP = {
        'CNV' : 0,
        'PCV': 0,
        'Non-wet-AMD' : 0,
    }
    print(phase, end='')
    for line in NEW_ALL[phase]:
        DISEASE_TEMP[line[1]] += 1
    print(' : ', DISEASE_TEMP['CNV'], DISEASE_TEMP['PCV'], DISEASE_TEMP['Non-wet-AMD'])
    DISEASE_TOTAL['CNV']+= DISEASE_TEMP['CNV']
    DISEASE_TOTAL['PCV']+= DISEASE_TEMP['PCV']
    DISEASE_TOTAL['Non-wet-AMD']+= DISEASE_TEMP['Non-wet-AMD']
print('total : ', DISEASE_TOTAL['CNV'], DISEASE_TOTAL['PCV'], DISEASE_TOTAL['Non-wet-AMD'])
```
